Data Cleaning

```python
dataframe.dropna(subset=['Attrition'], inplace=True)
```

```python
dataframe.fillna(dataframe.mean(numeric_only=True), inplace=True)
dataframe.fillna(dataframe.mode(), inplace=True)
```

```python
dataframe.drop_duplicates(inplace=True)
```

Creating input and output

```python
features = dataframe.drop('Attrition', axis=1)
labels = dataframe['Attrition']
```

```python
features
```

```python
labels
```

Conversion of features into numeric values

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
labels = encoder.fit_transform(labels)
labels
```

```python
features.nunique()
```

```python
all_columns = features.columns
for col in all_columns:
    if features[col].nunique() == 1:
        features = features.drop(columns=col)

features
```

```python
number_type = features.select_dtypes(include='number').columns
print(number_type)
```

```python
object_type = features.select_dtypes(include='object').columns
print(object_type)
```

```python
#print(features[object_type])
features[object_type] = features[object_type].astype('category')
features.dtypes
```

```python
object_binary = []
for obj in object_type:
    if features[obj].nunique() == 2:
        object_binary.append(obj)
for obj in object_binary:
    object_type = object_type.drop(obj)
object_type
```

```python
features[object_binary]
```

```python
for obj in object_binary:
    features[obj] = encoder.fit_transform(features[obj])
features
```

```python
features.dtypes
```

```python
features = pd.get_dummies(features).astype(int)
features
```

```python
features.dtypes
```

Scaling the features

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
def scaling(features, scale_type='minmax'):
    numbers = features.select_dtypes(include='number').columns
    if scale_type == 'minmax':
        scaler = MinMaxScaler()
    elif scale_type == 'standard':
        scaler = StandardScaler()
    features = scaler.fit_transform(features[numbers])
    return features

features_scaled = scaling(features[number_type], 'minmax')
features_scaled
```

```python
features_nonnum = features.drop(columns=number_type)
features_nonnum
```

```python
features_scaled_df = pd.DataFrame(features_scaled, columns=number_type)
features_scaled_df
```

```python
features_df = pd.concat([features_scaled_df, features_nonnum], axis=1)
features_df
```

```python
labels_df = pd.DataFrame(labels, columns=['Attrition'])
labels_df
```

Correlation analysis

```python
labels_series = labels_df['Attrition']
correlations = features_df.corrwith(labels_series)
correlations
```

```python
corrs = correlations.abs().nlargest(20)
check_cols = corrs.keys().to_list()
features_check = features_df[check_cols]
features_check
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

label_arr = labels_df[labels_df.columns[0]].to_list()

X_train, X_test, y_train, y_test = train_test_split(features_check, label_arr, test_size=0.1, random_state=42)
#print(len(X_test))
y_test_df = pd.DataFrame(y_test, columns=['Attrition'])


clf = LogisticRegression()
print(X_train)
print(y_train)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

y_pred_df = pd.DataFrame(y_pred)

#print(y_pred)
# print(len(y_pred))

ff = []

ff.append(y_pred_df)
ff.append(y_pred_df)

votes = pd.concat(ff, axis='columns')
predictions = votes.mode(axis='columns').to_numpy()
#print(predictions.flatten())

accuracy = accuracy_score(y_test, predictions.flatten())
print(f"Accuracy of Logistic Regression Classifier: {accuracy:.2f}")
```
